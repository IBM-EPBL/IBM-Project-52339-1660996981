#### 6) Finding and replacing the outliers

```python
sns.boxplot(df,orient='h')
# from this you can see all the function has the outlier we need to make the values lie between lower extreme value and upper extreme value
```

```python
# quantile
qnt=df.quantile(q=(0.75,0.25))
qnt
```

```python
# calculating thre upper and lower limit using the below formula
# upper extreme =Q3 + 1.5*IQR
# lower extreme =Q1-1.5*IQR
# IQR =  Q3-Q1
# Q3 is quantile of 0.75
# Q1 is quantile of 0.25
```

```python
IQR=qnt.loc[0.75]-qnt.loc[0.25]
IQR
```

```python
# upper extreme
u_e=qnt.loc[0.75]+1.5*IQR
u_e
```

```python
# lower extreme
l_e=qnt.loc[0.25]-1.5*IQR
l_e
```

```python
# replacing the outlier
for col in list(df.columns)[1:]:
        df[col]=np.where(df[col]>u_e[col]-0.2,df[col].mean(),df[col])
        df[col]=np.where(df[col]<l_e[col]+0.2,df[col].mean(),df[col])
```

```python
sns.boxplot(df,orient='h')
# Here you can see the outlier is replaced with the mean value
```

#### 7) Converting the categorical columns and performing encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l_encod=LabelEncoder()
df.Sex=l_encod.fit_transform(df.Sex)
df.head(10)
```

#### 8) spliting the dependent and independent variables

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

#### 9) Scale the independent variables

```python
from sklearn.preprocessing import scale
```

```python
x=scale(x)
x
```

```python
print(x.mean(),x.std())
# you can see that mean is nearly to zero and standard deviation is nearly to 1. this means data is scaled successfully
```

#### 10) Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(x_train.shape,y_train.shape)
```

```python
print(x_test.shape,y_test.shape)
```

#### 11) Building the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
l_r=LinearRegression()
l_r.fit(x_train,y_train)
```

#### 12) Train the model

```python
y_pred_train=l_r.predict(x_train)
```

```python
y_pred_train
```

#### 13) Test the Model

```python
y_pred_test=l_r.predict(x_test)
y_pred_test
```

```python
# predicting with random value
l_r.predict([[2,0.5,0.4,0.1,0.3,0.23,0.13,0.15]])
```

#### 14) Metrics analysis

```python
from sklearn import metrics
```

```python
r2_score=metrics.r2_score(y_test,y_pred_test)
r2_score
```

```python
err=np.sqrt(metrics.mean_squared_error(y_test,y_pred_test))
err
```
